```python
median_age = df["Age"].median()
df["Age"] = df["Age"].fillna(median_age)
```

```python
most_common = df["Embarked"].mode()[0]
df["Embarked"] = df["Embarked"].fillna(most_common)
```

```python
df = df.drop(columns=["Cabin"])
```

```python
df = df.drop_duplicates()
```

```python
df.groupby("Sex")["Survived"].mean()
```

```python
df.groupby("Pclass")["Survived"].mean()
```

```python
df.groupby("Pclass")["Age"].mean()
```

```python
df["Age_Category"] = pd.cut(df["Age"],
                           bins=[0, 10, 18, 30, 50, 100],
                           labels=["Kid", "Teen", "Young", "Middle", "Old"])

df.groupby("Age_Category")["Survived"].mean()
```

```python
df.query('Sex == "female" and Survived == 1')
```

```python
df[(df["Age"] <= 10) & (df["Survived"] == 1)]
```

```python
df[(df["Pclass"] == 1) & (df["Survived"] == 1)]
```

Insignts

Female passengers had a higher survival rate compared to males. First-class passengers were more likely to survive than lower classes. Children had better survival chances, indicating priority during rescue. The highest survival combination was female passengers in first class.
